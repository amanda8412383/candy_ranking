```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np



import statsmodels.api as sm
```

```python
df = pd.read_csv("candy-data.csv", index_col=0)
df.head()
```

```python
x0 = df.loc[:, ~df.columns.isin(['winpercent'])]
y = df[['winpercent']]
x = sm.add_constant(x0)	
results = sm.OLS(y,x).fit()
results.summary()
```

```python
df["choco_peanut"]  = df["chocolate"] * df["peanutyalmondy"]
df["choco_wafer"] = df["chocolate"] * df["crispedricewafer"]
df["choco_caramel"] = df["chocolate"] * df["caramel"]
```

```python
x1 = df.loc[:, ~df.columns.isin(['winpercent'])]
y = df[['winpercent']]
x = sm.add_constant(x1)	
results = sm.OLS(y,x).fit()
results.summary()
```

```python
df.drop(columns=['choco_wafer', 'choco_caramel'], inplace=True)
df["choco_peanut_wafer"] = df["chocolate"] * df["peanutyalmondy"] * df["crispedricewafer"]
df["choco_peanut_caramel"] = df["chocolate"] * df["peanutyalmondy"] * df["caramel"]
```

```python
x2 = df.loc[:, ~df.columns.isin(['winpercent'])]
y = df[['winpercent']]
x = sm.add_constant(x2)	
results = sm.OLS(y,x).fit()
results.summary()
```

```python
df.drop(columns=['choco_peanut_wafer', 'choco_peanut_caramel', 'choco_peanut'], inplace=True)
df["wafer_fruit"] = df["crispedricewafer"] * df["fruity"] 
df["wafer_nougat"] = df["crispedricewafer"] * df["nougat"] 
df["wafer_caramel"] = df["crispedricewafer"] * df["caramel"] 
df["wafer_peanutyalmondy"] = df["crispedricewafer"] * df["peanutyalmondy"] 
```

```python
x3 = df.loc[:, ~df.columns.isin(['winpercent'])]
y = df[['winpercent']]
x = sm.add_constant(x3)	
results = sm.OLS(y,x).fit()
results.summary()
```

```python
df.drop(columns=['wafer_fruit', 'wafer_nougat', 'wafer_caramel', 'wafer_peanutyalmondy'], inplace=True)
df["chocolate_peanutyalmondy"] = df["chocolate"] * df["peanutyalmondy"] 
df["wafer_chocolate_caramel"] = df["crispedricewafer"] * df["chocolate"] * df["caramel"]
```

```python
x4 = df.loc[:, ~df.columns.isin(['winpercent'])]
y = df[['winpercent']]
x = sm.add_constant(x4)	
results = sm.OLS(y,x).fit()
results.summary()
```
